## This notebook merges the two datasets (formal and informal) and coverts from CSV to parquet file.

```python
import pandas as pd
#%pip install fastparquet
import fastparquet
```

```python

# for use to combine formal and informal files - results from 1.1 
df1 = pd.read_csv("Nairobi_formal_50m_trees.csv")
df2 = pd.read_csv("Nairobi_informal_50m_trees.csv")
frames = [df1, df2]
result = pd.concat(frames)
result.to_parquet('Kenya_mock_training_data_wTrees.parquet', engine='fastparquet')
```
